# Expense Tracker with Input Validation

Full Stack Internship - Python Task

```
# ==========================================
# Expense Tracker with Input Validation
# ==========================================

print("===================================")
print("         Expense Tracker")
print("===================================")

total = 0

while True:

    try:
        expense = float(input("Enter Expense Amount (Rs.): "))

        if expense < 0:
            print("Expense cannot be negative.")
            continue

        total = total + expense

    except ValueError:
        print("Please enter a valid number.")
        continue

    choice = input("Add another expense? (yes/no): ").lower()

    if choice == "no":
        break

print("\n===================================")
print(f"Total Spent : Rs. {total:.2f}")
print("===================================")
print("Program Finished Successfully.")
```
