### Key Skills Covered:
- **Lists** (`append()`, `pop()`)
- **Loops & Control Flow** (`while`, `for`, `if-elif-else`)
- **Helper Functions** (`enumerate()`)
- **Input Validation & Exception Handling** (`try-except`, `.strip()`)

---
### Project Overview:
This is a feature-rich, console-based To-Do List application built using Python. It allows users to dynamically add tasks, view current tasks with proper indexing, and delete completed or unwanted tasks. The program includes input validation to prevent accidental crashes, ensuring a professional and user-friendly experience.

```python
def main():
    # Khali list jisme tasks store honge
    todo_list = []
    
    # Beautiful Header
    print("=" * 45)
    print("     WELCOME TO YOUR SMART TO-DO LIST      ")
    print("=" * 45)
    
    while True:
        # User Menu
        print("\n--- MENU OPTIONS ---")
        print("1. ➕ Add a New Task")
        print("2. 📋 View All Tasks")
        print("3. ❌ Delete a Task")
        print("4. 🚪 Exit Program")
        print("-" * 20)
        
        # User se choice lena
        choice = input("Enter your choice (1-4): ").strip()
        
        # 1. Task Add karne ke liye
        if choice == '1':
            task = input("\nEnter the task description: ").strip()
            if task:  # Check karna ke input khali na ho
                todo_list.append(task)
                print(f"✔️ Success: '{task}' has been added to your list!")
            else:
                print("⚠️ Error: Task cannot be empty!")
                
        # 2. Tasks View karne ke liye (Loop use karte hue)
        elif choice == '2':
            print("\n" + "=" * 15 + " YOUR TASKS " + "=" * 15)
            if not todo_list:
                print("Your list is currently empty. Add some tasks first!")
            else:
                # Loop through the list with index numbers
                for index, task in enumerate(todo_list, start=1):
                    print(f"{index}. [ ] {task}")
            print("=" * 42)
            
        # 3. Extra Feature: Task Delete karne ke liye (Bonus Marks!)
        elif choice == '3':
            if not todo_list:
                print("\n⚠️ No tasks to delete!")
                continue
                
            print("\n" + "=" * 15 + " CURRENT TASKS " + "=" * 15)
            for index, task in enumerate(todo_list, start=1):
                print(f"{index}. {task}")
            print("=" * 45)
            
            try:
                task_num = int(input("Enter the task number to delete: "))
                if 1 <= task_num <= len(todo_list):
                    removed_task = todo_list.pop(task_num - 1)
                    print(f"🗑️ Removed: '{removed_task}' has been deleted!")
                else:
                    print("⚠️ Invalid task number!")
            except ValueError:
                print("⚠️ Please enter a valid number!")
                
        # 4. Exit Program
        elif choice == '4':
            print("\nThank you for using Smart To-Do List! Goodbye! 👋")
            print("=" * 45)
            break
            
        else:
            print("⚠️ Invalid Choice! Please enter a number between 1 and 4.")

# Program ko run karne ke liye
if __name__ == "__main__":
    main()
```
